# TiRex Time Series Prediction Playbook

This playbook can be run easily after setting up the repository, no hardware requirements needed. (Only tested on Windows)

```python
def extend_path():
    """Extend notebooks system path config to import relative packages."""
    import sys
    from pathlib import Path

    parent_folder = str(Path.cwd().parent)
    print(f"Adding {parent_folder} to sys.path")
    if parent_folder not in sys.path:
        sys.path.insert(0, parent_folder)


extend_path()
```

```python
import torch
from tirex import load_model, ForecastModel
from playground.utils import plot_forecast
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

MODEL_PATH = "NX-AI/TiRex"
N_TIMESERIES = 1
```

```python
model: ForecastModel = load_model(path=MODEL_PATH, device=str(DEVICE))  # type: ignore
```

```python
context = torch.rand((N_TIMESERIES, 128))  # Sample Data (5 time series with length 128)
```

```python
quantiles, mean = model.forecast(context=context, prediction_length=64)
```

```python
plot_forecast(
    ctx=context,
    quantile_fc=quantiles,  # type: ignore
    real_future_values=None,
)
```
